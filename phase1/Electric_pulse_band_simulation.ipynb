# Two-Band Model with Time-Dependent Electric Field


## 1. Finding The Dipole Expectation Value

The purpose of this simulation is to find the dipole expectation value $\langle D(\vec{k}, t) \rangle$. The system is subject to an electric pulse, which impacts the dipole expectation value (DEV). 

The interband dipole expectation value over time is given by:

$$
\langle D(\vec{k}, t) \rangle = \text{Tr}\left[\rho(\vec{k}, t) D(\vec{k}) \right]
$$



In order to find the DEV, it is necessary to find the density matrix $ \rho(\vec{k}, t) $. 

The density matrix evolves according to :

$$
\frac{d}{dt} \rho(\vec{k}, t) = -i \left[ h(\vec{k}, t), \rho(\vec{k}, t) \right]
$$

### Density Matrix

The density matrix is:

$$
\rho(\vec{k}, t) =
\begin{bmatrix}
\rho_{VB} & c_1 + i c_2 \\
c_1 - i c_2 & \rho_{CB}
\end{bmatrix}
$$

Where:

- $ \rho_{VB}, \rho_{CB} $: occupations of the valence and conduction bands  
- $ c_1, c_2 $: real and imaginary parts of the interband coherence

Initial condition:

$$
\rho(\vec{k}, 0) =
\begin{bmatrix}
1 & 0 \\
0 & 0
\end{bmatrix}
$$




### Hamiltonian

The Hamiltonian of the system is required to find the time dependent density matrix. It is given by:

$$
h(\vec{k}, t) = h_0(\vec{k}) - \vec{E}(t) \cdot \vec{D}(\vec{k})
$$

with the unperturbed Hamiltonian:

$$
h_0(\vec{k}) =
\begin{bmatrix}
\varepsilon_{VB}(\vec{k}) & 0 \\
0 & \varepsilon_{CB}(\vec{k})
\end{bmatrix}
$$

and the dipole interaction:

$$
\vec{E}(t) \cdot \vec{D}(\vec{k}) = E(t)
\begin{bmatrix}
0 & D(\vec{k}) \\
D(\vec{k}) & 0
\end{bmatrix}
$$

So the full Hamiltonian becomes:

$$
h(\vec{k}, t) =
\begin{bmatrix}
\varepsilon_{VB}(\vec{k}) & -E(t) D(\vec{k}) \\
- E(t) D(\vec{k}) & \varepsilon_{CB}(\vec{k})
\end{bmatrix}
$$


#### The Electric Field

The electric field is assumed to have to form of a gaussian pulse:

$$
E(t) = \frac{E_0}{\sqrt{2 \pi \sigma^2}} e^{\frac{-(t - t_0)^2}{2 \sigma^2}}
$$

Where $\sigma$ is a tuning parameter for the width of the pule. 


#### The Band Structure

The band structure is defined as follows:

$$
\varepsilon_{VB}(k) = 2J \cos(ka) - 2J - \frac{\Delta}{2}
$$

$$
\varepsilon_{CB}(k) = -2J \cos(ka) + 2J + \frac{\Delta}{2}
$$


# 2. The time evolution of the density matrix at the heart of the problem. 

The density matrix encapsulates the whole dynamics of the problem. We will now focus on how to solve it. 

We study the time evolution of the density matrix $ \rho(\vec{k}, t) $ under a time-dependent electric field using the equation:

$$
\frac{d}{dt} \rho(\vec{k}, t) = -i \left[ h(\vec{k}, t), \rho(\vec{k}, t) \right]
$$

---

### Define the Matrices

**Density matrix:**

Let $\rho_{12} = c_1 + i c_2$ and $\rho_{21} = c_1 - i c_2$. Then:

$$
\rho(\vec{k}, t) =
\begin{bmatrix}
\rho_{VB} & \rho_{12} \\
\rho_{21} & \rho_{CB}
\end{bmatrix}
$$

**Hamiltonian:**

Let $V = -E(t) D(\vec{k})$. Then:

$$
h(\vec{k}, t) =
\begin{bmatrix}
\varepsilon_{VB}(\vec{k}) & V \\
V & \varepsilon_{CB}(\vec{k})
\end{bmatrix}
$$

---

### Compute the Commutator

Compute $h \rho$:

$$
h \rho =
\begin{bmatrix}
\varepsilon_{VB} \rho_{VB} + V \rho_{21} & \varepsilon_{VB} \rho_{12} + V \rho_{CB} \\
V \rho_{VB} + \varepsilon_{CB} \rho_{21} & V \rho_{12} + \varepsilon_{CB} \rho_{CB}
\end{bmatrix}
$$

Compute $\rho h$:

$$
\rho h =
\begin{bmatrix}
\rho_{VB} \varepsilon_{VB} + \rho_{12} V & \rho_{VB} V + \rho_{12} \varepsilon_{CB} \\
\rho_{21} \varepsilon_{VB} + \rho_{CB} V & \rho_{21} V + \rho_{CB} \varepsilon_{CB}
\end{bmatrix}
$$

Now compute the commutator $[h, \rho] = h \rho - \rho h$:


<!--
$$
[h, \rho] =
\begin{bmatrix}
V \rho_{21} - \rho_{12} V &
\Delta \varepsilon(\vec{k}) \rho_{12} + V (\rho_{CB} - \rho_{VB}) \\
\Delta \varepsilon(\vec{k}) \rho_{21} + V (\rho_{VB} - \rho_{CB}) &
V \rho_{12} - \rho_{21} V
\end{bmatrix}
$$

where:

$$
\Delta \varepsilon(\vec{k}) = \varepsilon_{CB}(\vec{k}) - \varepsilon_{VB}(\vec{k})
$$

---

### Step: Substituting $\rho_{12}$ and $\rho_{21}$

We substitute:

- $\rho_{12} = c_1 + i c_2$
- $\rho_{21} = c_1 - i c_2$

into the commutator:

$$
[h, \rho] =
\begin{bmatrix}
V(c_1 - i c_2) - (c_1 + i c_2)V &
\Delta \varepsilon(\vec{k})(c_1 + i c_2) + V (\rho_{CB} - \rho_{VB}) \\
\Delta \varepsilon(\vec{k})(c_1 - i c_2) + V (\rho_{VB} - \rho_{CB}) &
V(c_1 + i c_2) - (c_1 - i c_2)V
\end{bmatrix}
$$

---

### Step: Simplify each element

- Top-left element:

$$
V(c_1 - i c_2) - (c_1 + i c_2)V = V c_1 - i V c_2 - c_1 V - i c_2 V = [V, c_1] - i[V, c_2] = 2i V c_2
$$

(since $[V, c_1] = 0$ when $c_1$ is scalar)

- Bottom-right element:

$$
V(c_1 + i c_2) - (c_1 - i c_2)V = 2i V c_2
$$

- Off-diagonal elements stay as-is (no simplification needed yet):

---

### Final simplified commutator:

-->

$$
[h, \rho] =
\begin{bmatrix}
2i V c_2 &
\Delta \varepsilon(\vec{k})(c_1 + i c_2) + V (\rho_{CB} - \rho_{VB}) \\
\Delta \varepsilon(\vec{k})(c_1 - i c_2) + V (\rho_{VB} - \rho_{CB}) &
-2i V c_2
\end{bmatrix}
$$

where

$$
\Delta \varepsilon(\vec{k}) = \varepsilon_{CB}(\vec{k}) - \varepsilon_{VB}(\vec{k})
$$

---

Using: 
$$
\frac{d}{dt} \rho(\vec{k}, t) = -i \left[ h(\vec{k}, t), \rho(\vec{k}, t) \right]
$$ 

And: 
$$
\rho(\vec{k}, t) =
\begin{bmatrix}
\rho_{VB} & \rho_{12} \\
\rho_{21} & \rho_{CB}
\end{bmatrix}
$$

---

And get to:

$$
\begin{bmatrix}
\frac{d}{dt} \rho_{VB} & \frac{d}{dt} (c_1 + i c_2) \\
\frac{d}{dt} (c_1 - i c_2) & \frac{d}{dt} \rho_{CB}
\end{bmatrix}
=
\begin{bmatrix}
2 V c_2 &
-i \Delta \varepsilon(\vec{k})(c_1 + i c_2) - i V (\rho_{CB} - \rho_{VB}) \\
-i \Delta \varepsilon(\vec{k})(c_1 - i c_2) - i V (\rho_{VB} - \rho_{CB}) &
-2 V c_2
\end{bmatrix}
$$

# 3. Recasting the problem as a Linear Algebra problem

Define the state vector:

$$
\vec{y}(t) =
\begin{bmatrix}
\rho_{VB} \\
\rho_{CB} \\
\rho_{12} \\
\rho_{21}
\end{bmatrix}
$$

Then the system evolves as:

$$
\frac{d}{dt} \vec{y}(t) =
\begin{bmatrix}
0 & 0 & -iV & iV \\
0 & 0 & iV & -iV \\
-iV & iV & -i \Delta \varepsilon(\vec{k}) & 0 \\
iV & -iV & 0 & -i \Delta \varepsilon(\vec{k})
\end{bmatrix}
\cdot
\vec{y}(t)
$$

with:

- $\rho_{12} = c_1 + i c_2$
- $\rho_{21} = c_1 - i c_2$
- $V = -E(t) D(\vec{k})$
- $\Delta \varepsilon(\vec{k}) = \varepsilon_{CB}(\vec{k}) - \varepsilon_{VB}(\vec{k})$


```python
#When using jupyter notebook, use ipympl (pip install ipympl)
%matplotlib widget

import numpy as np
from scipy.integrate import solve_ivp #use to compute the matrix time evolution
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python

# PARAMETERS:
def eps_vb(J, delta, k): return 2 * J * np.cos(k) - 2*J - delta / 2 #diagonal element of the static hamiltonian
def eps_cb(J, delta, k): return -2 * J * np.cos(k) + 2*J + delta / 2 #diagonal element of the static hamiltonian
def delta_epsilon(k, J, delta): return eps_cb(J, delta, k) - eps_vb(J, delta, k) #distance eps_cb and eps_cv at their closest
def D(k): return (k-3.14)*1.0  # constant dipole

# Electric pulse
def E(t, sigma, E0, t0=0.0):
    return E0 / (np.sqrt(2 * np.pi * sigma**2)) * np.exp((-(t - t0) ** 2) / (2 * sigma**2))

# V: the product of the electric pulse and D(k)
def V(t, k, sigma, E0): return -E(t, sigma, E0) * D(k)


# returns the 4D derivation matrix
def A(t, k, J, delta, sigma, E0):
    v = V(t, k, sigma, E0)
    dE = delta_epsilon(k, J, delta)
    return np.array([
        [0,     0,     -1j*v,   1j*v],
        [0,     0,      1j*v,  -1j*v],
        [-1j*v, 1j*v, -1j*dE,     0],
        [1j*v, -1j*v,     0,  -1j*dE]
    ], dtype=complex)

# returns the matrix product (=@) of the time derivative and the given vector
def rhs(t, y, k, J, delta, sigma, E0):
    return A( t, k, J, delta, sigma, E0) @ y




```

```python
# function to: 
# 1) compute the time evolution of the density matrix
# 2) plot the band structure and electric pulse
# 3) plot the time evolution of the system

def solve_and_plot(J, k, delta, sigma, E0):
    y0 = np.array([1.0, 0.0, 0.0, 0.0], dtype=complex) #initial condition for the density matrix
    t_span = (0, 20)
    t_eval = np.linspace(*t_span, 1000)
    sol = solve_ivp(lambda t, y: rhs(t, y, k, J, delta, sigma, E0), t_span, y0, t_eval=t_eval, method="RK45") #uses fourth order Runge Kutta to find the time evolution of the matrix

    #plot the Density matrix
    ax3.clear()
    ax3.plot(sol.t, sol.y[0].real, label=r'$Re(\rho_{VB})$')
    ax3.plot(sol.t, sol.y[1].real, label=r'$Re(\rho_{CB})$')
    ax3.set_xlabel("Time")
    ax3.set_ylabel("Population")
    ax3.set_title("Density Matrix Evolution")
    ax3.grid(True)
    ax3.legend()

    #plot the electric pulse
    ax2.clear()
    ax2.plot(t_eval, E(t_eval, sigma, E0), label='Electric Pulse E(t)')
    ax2.set_title("Electric Field Pulse")
    ax2.set_xlabel("Time")
    ax2.set_ylabel("E(t)")
    ax2.grid(True)
    ax2.legend()

    fig.canvas.draw_idle()
```

```python

# Using the functions from above: 
# Defining the initial parameters: 
J0 = 1.45
delta0 = 4.26
k0 = 0 #-0.8
sigma0 = 0.12
E0_0 = 0.4


# Set up the plot
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 10))
plt.subplots_adjust(left=0.25, bottom=0.45)

k_vals = np.linspace(-np.pi, np.pi, 200)
line1, = ax1.plot(k_vals, eps_vb(J0, delta0, k_vals), label="VB")
line2, = ax1.plot(k_vals, eps_cb(J0, delta0, k_vals), label="CB")
ax1.set_title("Band Structure")
ax1.legend()

solve_and_plot(J0, k0, delta0, sigma0, E0_0)


####### ----------- INTERACTIVE PLOTTING --------------

# define the sliders for interactive plot
ax_J = plt.axes([0.25, 0.35, 0.65, 0.03])
ax_k = plt.axes([0.25, 0.30, 0.65, 0.03])
ax_delta = plt.axes([0.25, 0.25, 0.65, 0.03])
ax_sigma = plt.axes([0.25, 0.20, 0.65, 0.03])
ax_E0 = plt.axes([0.25, 0.15, 0.65, 0.03])


slider_J = Slider(ax_J, 'J', -2*0.1, 2*3.0, valinit=J0)
slider_k = Slider(ax_k, 'k', -2*np.pi, 2*np.pi, valinit=k0)
slider_delta = Slider(ax_delta, '\u03b4', -2.0, 5.0, valinit=delta0)
slider_sigma = Slider(ax_sigma, '\u03c3', 0.01, 1.0, valinit=sigma0)
slider_E0 = Slider(ax_E0, 'E0', 0.001, 1.0, valinit=E0_0)


# Update function for interactive plot 
def update(val):
    J = slider_J.val
    k = slider_k.val
    delta = slider_delta.val
    sigma = slider_sigma.val
    E0 = slider_E0.val

    

    line1.set_ydata(eps_vb(J, delta, k_vals))
    line2.set_ydata(eps_cb(J, delta, k_vals))
    ax1.relim()
    ax1.autoscale_view()

    solve_and_plot(J, k, delta, sigma, E0)

slider_J.on_changed(update)
slider_k.on_changed(update)
slider_delta.on_changed(update)
slider_sigma.on_changed(update)
slider_E0.on_changed(update)


plt.show()
```

# DMD 

```python
from matplotlib.widgets import Slider
import matplotlib.pyplot as plt
import numpy as np

def interactive_full_system_dmd_plot(r, filename="density_vs_k.txt"):
    # Load and DMD
    def load_full_k_dmd_data(filename):
        data = np.loadtxt(filename)
        k_vals = np.unique(data[:, 0]) 
        t_vals = np.unique(data[:, 1])
        num_k = len(k_vals)
        num_t = len(t_vals)

        rho_vb = data[:, 2].reshape(num_k, num_t)
        rho_cb = data[:, 3].reshape(num_k, num_t)
        X_full = np.vstack([rho_vb, rho_cb])  

        return X_full[:, :-1], X_full[:, 1:], t_vals[:-1], rho_vb, rho_cb, k_vals


#Note: for a detailed procedure on DMD algorithm using SVD, see: 
## "Using Dynamic Model decomposition to Predict The Dynamics of a two-time Non-equilibrium Green's function", Yin et al. 
    def perform_dmd(X, Xp, t, r):
    
    # --------- Decomposition part of DMD:

        U, S, Vh = np.linalg.svd(X, full_matrices=False) #perform SVD on X_0
        #r = min(r, len(S))  #ensure the r is not larger than the rank
        U_r, S_r, Vh_r = U[:, :r], np.diag(S[:r]), Vh[:r, :] #computes the reduced SVD

        A_tilde = U_r.T @ Xp @ Vh_r.T @ np.linalg.inv(S_r) # Finds A_tilde = U^cross @ X_1 @ Vh_r^cross @ S^-1
        eigvals, W = np.linalg.eig(A_tilde) #finds the eigenvvalues and vectors of A_tilde

    # --------- Reconstruction part of DMD:
        Phi = Xp @ Vh_r.T @ np.linalg.inv(S_r) @ W # computes the "DMD modes"
        dt = t[1] - t[0]
        omega = np.log(eigvals) / dt #logarithm of the eigenvalues over the time step
        b = np.linalg.pinv(Phi) @ X[:, 0] #b=phi^t @ X_0

        #X_1 = Phi @ exp(Omega*t)*b
        time_dynamics = np.array([b * np.exp(omega * T) for T in t]).T
        X_dmd = Phi @ time_dynamics

        return eigvals, X_dmd.real

    # Load data and perform DMD
    X, Xp, t, rho_vb_orig, rho_cb_orig, k_vals = load_full_k_dmd_data(filename)
    eigvals, X_dmd = perform_dmd(X, Xp, t, r)
    num_k = len(k_vals)

    rho_vb_dmd = X_dmd[:num_k, :]
    rho_cb_dmd = X_dmd[num_k:, :]





## ---------- Plot setup ----------------


    fig, ax = plt.subplots(figsize=(8, 4))
    plt.subplots_adjust(bottom=0.25)

    i0 = 0
    line1, = ax.plot(t, rho_vb_orig[i0, :-1], label="Original VB")
    line2, = ax.plot(t, rho_vb_dmd[i0], '--', label="DMD VB")
    line3, = ax.plot(t, rho_cb_orig[i0, :-1], label="Original CB")
    line4, = ax.plot(t, rho_cb_dmd[i0], '--', label="DMD CB")

    ax.set_title(f"Full-System DMD Reconstruction at k-index {i0} (k = {k_vals[i0]:.3f})")
    ax.set_xlabel("Time")
    ax.set_ylabel("Population")
    ax.grid(True)
    ax.legend()

    # Slider
    ax_slider = plt.axes([0.2, 0.1, 0.6, 0.03])
    slider = Slider(ax_slider, "k index", 0, num_k - 1, valinit=i0, valstep=1)

    def update(val):
        i = int(slider.val)
        line1.set_ydata(rho_vb_orig[i, :-1])
        line2.set_ydata(rho_vb_dmd[i])
        line3.set_ydata(rho_cb_orig[i, :-1])
        line4.set_ydata(rho_cb_dmd[i])
        ax.set_title(f"Full-System DMD Reconstruction at k-index {i} (k = {k_vals[i]:.3f})")
        fig.canvas.draw_idle()

    slider.on_changed(update)
    plt.show()




```

```python
interactive_full_system_dmd_plot(25, "density_vs_k.txt") #first argument: set r for a more or less precise fitting
```
