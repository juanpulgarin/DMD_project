```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

import Hamiltonian as Hamiltonian
import Plotting as plotting
import constants as cst
```

### This section shows what happens when changing the different parameters in your Hamiltonian

```python
k_vals = np.linspace(-np.pi, np.pi, 101)[:-1]

J_list     = np.array([1.0])/cst.Ry
delta_list = np.array([0.5,1.5])/cst.Ry

valence_bands    = {}
conduction_bands = {}
for J in J_list:
    for  in delta_list:
        valence_bands[J,]    = Hamiltonian.eps_vb( k_vals, J,  )
        conduction_bands[J,] = Hamiltonian.eps_cb( k_vals, J,  )

limits=np.array([[-np.pi,np.pi],[-3.5,3.5]])

plotting.plot_band_structure(k_vals, J_list, delta_list, valence_bands, conduction_bands, limits,fout='')

for J in J_list:
    for  in delta_list:
        plotting.plot_band_structure(k_vals, np.array([J]), np.array([]), valence_bands, conduction_bands, limits,fout='')



J_list     = np.array([1.0,2.5])/cst.Ry
delta_list = np.array([0.5])/cst.Ry


valence_bands    = {}
conduction_bands = {}
for J in J_list:
    for  in delta_list:
        valence_bands[J,]    = Hamiltonian.eps_vb( k_vals, J,  )
        conduction_bands[J,] = Hamiltonian.eps_cb( k_vals, J,  )

limits=np.array([[-np.pi,np.pi],[-3.5,3.5]])

plotting.plot_band_structure(k_vals, J_list, delta_list, valence_bands, conduction_bands, limits,fout='')

for J in J_list:
    for  in delta_list:
        plotting.plot_band_structure(k_vals, np.array([J]), np.array([]), valence_bands, conduction_bands, limits,fout='')
```

### Fix a set of parameters for the Bands, then change the values of the Field and the case

```python
k_vals = np.linspace(-np.pi, np.pi, 101)[:-1]

J_list     = np.array([0.5])/cst.Ry
delta_list = np.array([0.5])/cst.Ry

valence_bands    = {}
conduction_bands = {}
for J in J_list:
    for  in delta_list:
        valence_bands[J,]    = Hamiltonian.eps_vb( k_vals, J,  )
        conduction_bands[J,] = Hamiltonian.eps_cb( k_vals, J,  )

limits=np.array([[-np.pi,np.pi],[-2.5,2.5]])

plotting.plot_band_structure(k_vals, J_list, delta_list, valence_bands, conduction_bands, limits,fout='')
```

```python
case=1

t = np.linspace(0, 100, 1000)/ cst.tfs

FWHM  = 20. / cst.tfs
sigma = FWHM/(2.0*np.sqrt(2*np.log(2)))
E0 = 1.7

efield = Hamiltonian.E(t, sigma,E0, =0.5/cst.Ry, case=case )
#plotting.plot_electric_pulse(t*cst.tfs, efield, np.array([[0,100],[-0.002*0,0.002]]),fout='')

J     = 0.5/cst.Ry
delta = 0.5/cst.Ry
k_list=[0, 0.25, 0.50, 0.75]


t_span= (t[0],t[-1])


solutions=np.zeros((len(k_list),2,len(t)),dtype=np.complex64)
for k_i,k in enumerate(k_list):
    y0 = np.array([1.0, 0.0, 0.0, 0.0], dtype=complex)
    sol = solve_ivp(lambda t, y: Hamiltonian.rhs(t, y, k, J, delta, sigma, E0, =0.5/cst.Ry,case=case), t_span, y0, t_eval=t)
    solutions[k_i,0,:] = sol.y[0]
    solutions[k_i,1,:] = sol.y[1]
```

```python
limits=np.array([[0,100],[0.0,0.2]])
plotting.plot_population_field(t*cst.tfs,efield,k_list,solutions.real,limits)
```
